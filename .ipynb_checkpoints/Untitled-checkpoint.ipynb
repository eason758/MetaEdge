```python
def derivative(x1, x2, z1, z2):
    return (z2 - z1) / (x2 - x1)
```

```python
def finiteDifferenceInformationGain(x1, y1, delta_x, delta_y, x_name, y_name, train_set):
    z1 = informationGain(x1, y1, x_name, y_name, train_set)
    z1_delta_x = informationGain(x1 + delta_x, y1, x_name, y_name, train_set)
    z1_delta_y = informationGain(x1, y1 + delta_y, x_name, y_name, train_set)
    print(f'z1: {z1}, z1_delta_x: {z1_delta_x}, z1_delta_y: {z1_delta_y}')
    
    dz_dx2 = (informationGain(x1 + delta_x * 2, y1, x_name, y_name, train_set) - 2 * z1_delta_x + z1) / (delta_x * 2)
    dz_dydx = (informationGain(x1 + delta_x, y1 + delta_y, x_name, y_name, train_set) - 2 * z1_delta_x + z1) / (delta_x + delta_y)
    dz_dy2 = (informationGain(x1, y1 + delta_y * 2, x_name, y_name, train_set) - 2 * z1_delta_y + z1) / (delta_y * 2)
    dz_dxdy = (informationGain(x1 + delta_x, y1 + delta_y, x_name, y_name, train_set) - 2 * z1_delta_y + z1) / (delta_y + delta_x)
    
    grad = np.array([(z1_delta_x - z1) / delta_x , (z1_delta_y - z1) / delta_y])
    hessian = np.array([[dz_dx2, dz_dxdy], [dz_dydx, dz_dy2]])
    print(f'grad: {grad}\nhessian:\n {hessian}')
    
    return grad, hessian
```

```python
def finiteDifferenceRecall(x1, y1, x2, y2, x_name, y_name, train_set):
    z1 = Recall(x1, y1, x_name, y_name, train_set)
    z1_delta_x = Recall(x1 + delta_x, y1, x_name, y_name, train_set)
    z1_delta_y = Recall(x1, y1 + delta_y, x_name, y_name, train_set)
    print(f'z1: {z1}, z1_dela_x: {z1_delta_x}, z1_delta_y: {z1_delta_y}')
    
    dz_dx2 = (Recall(x1 + delta_x * 2, y1, x_name, y_name, train_set) - 2 * z1_delta_x + z1) / (delta_x * 2)
    dz_dydx = (Recall(x1 + delta_x, y1 + delta_y, x_name, y_name, train_set) - 2 * z1_delta_x + z1) / (delta_x + delta_y)
    dz_dy2 = (Recall(x1, y1 + delta_y * 2, x_name, y_name, train_set) - 2 * z1_delta_y + z1) / (delta_y * 2)
    dz_dxdy = (Recall(x1 + delta_x, y1 + delta_y, x_name, y_name, train_set) - 2 * z1_delta_y + z1) / (delta_x + delta_y)
    
    grad = np.array([(z1_delta_x - z1) / delta_x, (z1_delta_y - z1) / delta_y])
    hessian = np.array([[dz_dx2, dz_dxdy], [dz_dydx, dz_dy2]])
    print(f'grad: {grad}\nhessian:\n{hessian}')
        
    return grad, hessian
```

```python
def Recall(x_thre, y_thre, x_name, y_name, train_set):
    new_sam_1, new_sam_0 = newSamClassification(x_thre, y_thre, x_name, y_name, train_set)
    TP = new_sam_1[new_sam_1[SAR_type] != 0].shape[0]
    TN = new_sam_0[new_sam_0[SAR_type] == 0].shape[0]
    FP = new_sam_1[new_sam_1[SAR_type] == 0].shape[0]
    FN = new_sam_0[new_sam_0[SAR_type] != 0].shape[0]
    
    return TP / (TP+FN)
```

```python
def newSamClassification(x_thre, y_thre, x_name, y_name, train_set):
    idx1 = set(train_set[train_set[x_name] > x_thre].index)
    idx2 = set(train_set[train_set[x_name] > y_thre].index)
    idx = list(idx1 & idx2)
    
    new_sam_1 = train_set.iloc[idx]
    new_sam_0 = train_set.drop(index= new_sam_1.index)
    
    if len(set(new_sam_1.index) & set(new_sam_0.index)) != 0:
        raise ValueError('new sam 0 and new sam 1 are overlapped')
    
    return new_sam_1, new_sam_0
```

```python
def conditionalEntropy(new_sam_0, new_sam_1):
    #print(new_sam_0.shape[0], new_sam_1.shape[0])
    total = len(new_sam_0) + len(new_sam_1)
    condition_entropy = Entropy(new_sam_0, SAR_type) * (len(new_sam_0) / total) \
                        + Entropy(new_sam_1, SAR_type) * (len(new_sam_1) / total)
    
    return condition_entropy
```

```python
def informationGain(x_thre, y_thre, x, y, train_set):
    idx1 = set(train_set[train_set[x] >= x_thre].index)
    idx2 = set(train_set[train_set[y] >= y_thre].index)
    
    idx = list(idx1 & idx2)
    new_sam_1 = train_set.iloc[idx]
    new_sam_0 = train_set.drop(index= new_sam_1.index)
    
    if len(set(new_sam_0.index) & set(new_sam_1.index)) != 0:
        raise ValueError('new sam 0 and new sam 1 are overlapped')
    
    #print(f'new sam 1: {new_sam_1.shape[0]}, new sam 0: {new_sam_0.shape[0]}')
    information_gain = -(father_entropy - conditionalEntropy(new_sam_0, new_sam_1))
    
    return information_gain
```

```python
def CalRecallFilterRate(x_thre, y_thre, x, y, train_set):
    idx1 = set(train_set[train_set[x] >= x_thre].index)
    idx2 = set(train_set[train_set[y] >= y_thre].index)
    
    idx = list(idx1 & idx2)
    
    new_sam_1 = train_set.iloc[idx]
    new_sam_0 = train_set.drop(index= new_sam_1.index)
    
    if len(set(new_sam_0.index) & set(new_sam_1.index)) != 0:
        raise ValueError('new sam 0 and new sam 1 are overlapped!')
        
    TP = new_sam_1[new_sam_1[SAR_type] != 0].shape[0]
    TN = new_sam_0[new_sam_0[SAR_type] == 0].shape[0]
    FP = new_sam_1[new_sam_1[SAR_type] == 0].shape[0]
    FN = new_sam_0[new_sam_0[SAR_type] != 0].shape[0]
    
    recall = TP / (TP+FN)
    filter_rate = (TN + FN) / (TP + TN + FP + FN)
    
    return recall, filter_rate
```

```python
def initialPoint(x_name, y_name, train_set):
    x_median = SplitMedian(train_set, x_name)
    y_median = SplitMedian(train_set, y_name)
    
    # x_point1, y_point1 = x_median[np.random.randint(0, len(x_median))], x_median[np.random.randint(0, len(y_median))]
    # y_point1, y_point2 = y_median[np.random.randint(0, len(y_median))], y_median[np.random.randint(0, len(y_median))]
    
    x_random = np.random.randint(0, len(x_median))
    y_random = np.random.randint(0, len(y_median))
    x_point1, x_point2 = x_median[x_random], x_median[x_random + 1]
    y_point1, y_point2 = y_median[y_random], y_median[y_random + 1]
    
    print(f'x_point1: {x_point1}, y_point1: {y_point1}')
    print(f'x_point2: {x_point2}, y_point2: {y_point2}')
    return  x_point1, x_point2, y_point1, y_point2
```

```python
# V1 輔助計算斜率的點是最近的大於的中位數
def Newton(x1, x2, y1, y2, x_name, y_name, n_max, error):
    results = pd.DataFrame(columns= ['iterations number' ,'x1', 'y1', 'x2', 'y2', 'new_x1', 'new_y1', 'information gain', 'recall', 'filter rate'])
    n = 1
    while True:
        if n < n_max:
            z1 = informationGain(x1, y1, x_name, y_name, train_set)
            z2 = informationGain(x2, y2, x_name, y_name, train_set)
            print(f'information gain z1: {z1}, z2: {z2}')
            
            grads = np.array([derivative(x1, x2, z1, z2), derivative(y1, y2 ,z1, z2)])
            print(f'gradients: {grads}')
            if np.count_nonzero(grads) != len(grads):
                print(f'finish reason: gradient equal zero ')
                break
                
            new_x1 = x1 - learning_rate * (z1 / grads[0])
            new_y1 = y1 - learning_rate * (z1 / grads[1])
            
            information_gain = informationGain(new_x1, new_y1, x_name, y_name, train_set)
            recall, filter_rate = CalRecallFilterRate(new_x1, new_y1, x_name, y_name, train_set)
            
            
            results = results.append({'iterations number': n, 'x1': x1, 'y1': y1, 'x2': x2, 'y2': y2, 'new_x1': new_x1, 'new_y1': new_y1, \
                                    'information gain': information_gain, 'recall': recall, 'filter rate': filter_rate}, ignore_index= True)
            
            x1 = new_x1
            y1 = new_y1
            # x2 = new_x1 + 1
            # y2 = new_y1 + 1
            
            print(f'iterations number: {n}, new x1: {new_x1}, new y1: {new_y1}')
            x2 = x_median[x_median > new_x1][0]
            y2 = y_median[y_median > new_y1][0]
            
            if abs(informationGain(new_x1, new_y1, x_name, y_name, train_set) - informationGain(x1, y1, x_name, y_name, train_set)) < error:
                print(f'finish reason: error is enough small\nfinal terations number: {n}, new x1: {new_x1}, new y1: {new_y1}')
                break
            
            
            n += 1
        else:
            print('finish reason: n == n_max')
            break
    return results
```

```python
# hyperparameter
error = 0
n_max = 100
SAR_type = 'TO_SAR'
father_entropy = Entropy(train_set, SAR_type)
delta_x = (train_set['Credit_Amt'].max() - train_set['Credit_Amt'].min()) / len(train_set)
delta_y = (train_set['number_of_Credit'].max() - train_set['number_of_Credit'].min()) / len(train_set)
learning_rate = 1e-3

print(f'father_entropy: {father_entropy}, delta_x: {delta_x}, delta_y: {delta_y}')
```

```python
# V2 加入二階微分的部分，輔助計算的點是用個別全距除的平均
def NewtonV2(x1, x2, y1, y2, x_name, y_name, n_max, error):
    results = pd.DataFrame(columns= ['iterations number' ,'x1', 'y1', 'new_x1', 'new_y1', 'information gain', 'recall', 'filter rate'])
    n = 1
    while True:
        if n < n_max:
            grads, hessian = finiteDifferenceInformationGain(x1, y1, delta_x, delta_y, x_name, y_name, train_set)
            print(f'gradients: {grads}')
            new_point = np.array([x1, y1]) - np.linalg.inv(hessian) @ grads
            
            
            
            information_gain = informationGain(new_point[0], new_point[1], x_name, y_name, train_set)
            recall, filter_rate = CalRecallFilterRate(new_point[0], new_point[1], x_name, y_name, train_set)
            
            
            results = results.append({'iterations number': n, 'x1': x1, 'y1': y1, 'new_x1': new_point[0], 'new_y1': new_point[1], \
                                    'information gain': information_gain, 'recall': recall, 'filter rate': filter_rate}, ignore_index= True)
            
            x1 = new_point[0]
            y1 = new_point[1]
            # x2 = new_x1 + 1
            # y2 = new_y1 + 1

            #x2 = x_median[x_median > x1][0]
            #y2 = y_median[y_median > y1][0]
            
            if abs(informationGain(new_point[0], new_point[1], x_name, y_name, train_set) - informationGain(x1, y1, x_name, y_name, train_set)) < error:
                print(f'finish reason: error is enough small\nfinal terations number: {n}, new x1: {new_x1}, new y1: {new_y1}')
                break
            
            print(f'iterations number: {n}, new x1: {new_point[0]}, new y1: {new_point[1]}')
            n += 1
        else:
            print('finish reason: n == n_max')
            break
    return results
```
