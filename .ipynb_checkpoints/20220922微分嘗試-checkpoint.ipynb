# 用 tensorflow 做微分

```python
TRAIN_STEP = 25

train_x1 = np.linspace(-1, 1, 100, dtype= 'float32')
train_x2 = np.linspace(1, 2, 100, dtype= 'float32')
train_y = 2 + tf.pow(train_x1, 2) + tf.pow(train_x2, 4) + np.random.random(*train_x1.shape) * 0.33 + 10

print(train_x1.shape, train_x2.shape, train_y.shape)

w1 = tf.Variable(0.0, dtype= 'float32')
w2 = tf.Variable(0.0, dtype= 'float32')
b = tf.Variable(1.0, dtype= 'float32')

optimizer = tf.keras.optimizers.SGD(learning_rate= 0.01)
mse = tf.keras.losses.MeanSquaredError()

for i in range(TRAIN_STEP):
    print('epoch: ', i)
    print('w1: ', w1.numpy())
    print('w2: ', w2.numpy())
    print('b: ', b.numpy())
    
    with tf.GradientTape() as tape:
        y_hat = w1 * tf.pow(train_x1, 2) + w2 * tf.pow(train_x2, 4) + b
        loss = mse(train_y, y_hat)
    gradients = tape.gradient(target= loss, sources= [w1, w2, b])
    print('gradient:', gradients)
    optimizer.apply_gradients(zip(gradients, [w1, w2, b]))
    
plt.figure()
plt.plot(train_x1, train_y, 'r+')
plt.plot(train_x1, w1 * tf.pow(train_x1, 2) + w2 * tf.pow(train_x2, 4) + b)
plt.show()

plt.figure()
plt.plot(train_x2, train_y, 'g+')
plt.plot(train_x2, w1 * tf.pow(train_x1, 2) + w2 * tf.pow(train_x2, 4) + b)
```

```python
def Enrtopy(data, feature):
    total_cnt = len(data)
    if total_cnt == 0:
        return 0
    else:
        target0_cnt = len(data[data[feature] == 1])
        target1_cnt = len(data[data[feature] == 0])
        
        if (target0_cnt == 0) | (target1_cnt == 1):
            entropy = 0
        else:
            entropy = -((target0_cnt / total_cnt) * log2(target0_cnt / total_cnt) + (target1_cnt / total_cnt) * log2(target1_cnt / total_cnt))
        
        return entropy
```

# 對 information gain 做微分
D 是資料集，A是分裂條件，k 是類別這裡是分兩類，i 則是由 feature A 將 D 劃分成幾個子集<br>
$information gain IG(D,A)= H(D) - H(D|A=a)$ <br>
這裡的機率是根據資料集的樣本數，計算出來 <br>
$H(D)= -\sum_{k=1}{2}\frac{|C_k|}{D}log_{2}\frac{|C_k|}{|D|}$
$H(D|A)= \sum_{i=1}^{2}\frac{|D_{i}|}{|D|}\sum_{k=1}^{2}\frac{|D_{ik}|}{|D_i|}log_{2}\frac{|D_ik|}{|D_i|}$

```python
credit_amt_median = np.percentile(credit_amt_median, np.linspace(10, 100, 50))
number_of_credit_median = np.percentile(number_amt_median, np.linspace(10, 100, 50))
```

```python
results = pd.DataFrame(columns= ['credit_amt_median', 'number_of_credit_median', 'prob0', 'prob1'])
total_cnt = len(train_set) - 1

for thre1 in range(len(credit_amt_median)):
    for thre2 in range(len(number_amt_median)):
        idx1 = set(train_set[train_set['Credit_Amt'] >= thre1].index)
        idx2 = set(train_set[train_set['number_of_Credit'] >= thre2].index)
        
        idx = list(idx1 & idx2)
        new_sam_1 = train_set.iloc[idx]
        new_sam_0 = train_set.drop(index= new_sam_1.index)
        
        if len(set(new_sam_0.index) & set(new_sam_1.index)) != 0:
            raise ValueError('new sam 0 and new sam 1 are overlapped!')
        
        if len(new_sam_0) == 0 or len(new_sam_1) == 0:
            continue
            
        prob0 = len(new_sam_0[new_sam_0['TO_SAR']== 0]) / len(new_sam_0)
        prob1 = len(new_sam_1[new_sam_1['TO_SAR']== 0]) / len(new_sam_1)
        
        results = results.append({'credit_amt_median' : thre1, 'number_of_credit_median' : thre2, 'prob0' : prob0, 'prob1': prob1}, ignore_index= True)
```
