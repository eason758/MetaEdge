```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ \frac{1}{1+e^{-(b + wx)}}$$

```python
def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))
```

```python
b = 0.8
```

```python
def sigmoid1(x):
    return 1.0 / (1 + np.exp(-(b + 5 * x)))
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
y1 = sigmoid1(x)
y2 = sigmoid1(0.8)
print(y2)
```

```python
import math
math.log(1e-10)
```

```python
plt.figure()
plt.plot(x, y, label= 'sigmoid')
plt.plot(x, y1, label= 'sigmoid shift')
plt.axvline(0.8, c= 'r')
plt.plot(0.8, y2, 'go')
plt.grid()
plt.ylim(0, 1)
plt.xlim(-10, 10)
plt.legend()
plt.title('sigmoid function')
plt.savefig('sigmoid.png')
plt.show()
```
