```python
conn = pymssql.connect(server='.', user='NCTUUser2', password='NCTU@MetaEdge', database='NCTU(1090401-1091231)')
cursor = conn.cursor()
```

```python
START = datetime.datetime.now()
count=0
for data in days_df:
    clear_output(wait=True)
    for i in range(data.shape[0]):
        print('row:{}, day{}'.format(i, days[count]))
        run_date = data['run_date'][i].replace(" 00:00:00", "").replace("-", "")
        run_date = datetime.datetime.strptime(run_date, "%Y%m%d")

        row_index = []
        skip_row_index = []

        cust_code = mapping[data['Cust_No'][i]]
        file_path = origin_path + str(cust_code) + end_of_file

        if os.path.isfile(file_path):
            cust_txn = pd.read_csv(file_path, usecols = ['Txn_Dt_Key', 'Acct_No'])

            # search run date第前x個營業日的日期
            cursor.execute("DECLARE @CYCLE_END DATE =(%s)\
                SELECT DATADT FROM(\
                SELECT ROW_NUMBER() OVER (ORDER BY DATADT DESC) AS [ROW]\
                , DATADT\
                FROM [NCTU(1090401-1091231)].[dbo].[CB_DT]\
                WHERE DATADT <= @CYCLE_END\
                AND HOLIDY_FG =0\
                AND Contry2Cd ='TW')S\
                WHERE [ROW] <=(%s) ",(run_date, days[count]))

            for row in cursor:
                before_run_date = row[0]

            for j in range(cust_txn.shape[0]):
                txn_dt_key = datetime.datetime.strptime(str(cust_txn['Txn_Dt_Key'][j]),"%Y%m%d")
                if txn_dt_key <= run_date and txn_dt_key >= before_run_date:
                    row_index.append(j)
                else:
                    skip_row_index.append(j)

            print(before_run_date, run_date)
            skip_row_index = [i+1 for i in skip_row_index]
            res = pd.read_csv(file_path,skiprows=skip_row_index)

            if not res.empty:
                print('save')
                row_txn = pd.DataFrame(res.values, columns = txn_col)
                row_txn = row_txn.sort_values(by=['Txn_Dt_Key'])

                first_date = datetime.datetime.strptime(str(row_txn.iloc[0]['Txn_Dt_Key']),"%Y%m%d")
                last_date = datetime.datetime.strptime(str(row_txn.iloc[-1]['Txn_Dt_Key']),"%Y%m%d")
                #print(first_date, last_date, run_date, days[count])

                if first_date < before_run_date or last_date > run_date:
                    raise ValueError("date mismatch: txn data earliear than 20days before or later than run date!")

                row_txn.to_csv(dest_path + str(days_name[count]) +'/'+ str(i) + end_of_file, index = False)
        else:
            print('{}_txn is empty'.format(cust_code))

    count+=1
    
print(f'total time: {datetime.datetime.now()-START}')
```
